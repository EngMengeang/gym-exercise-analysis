## Dataset Overview
**973 rows × 15 columns**

| Column | Description |
|--------|-------------|
| Age | Age of the gym member |
| Gender | Gender of the gym member (Male or Female) |
| Weight (kg) | Member's weight in kilograms |
| Height (m) | Member's height in meters |
| Max_BPM | Maximum heart rate (beats per minute) during workout sessions |
| Avg_BPM | Average heart rate during workout sessions |
| Resting_BPM | Heart rate at rest before workout |
| Session_Duration (hours) | Duration of each workout session in hours |
| Calories_Burned | Total calories burned during each session |
| Workout_Type | Type of workout performed (e.g., Cardio, Strength, Yoga, HIIT) |
| Fat_Percentage | Body fat percentage of the member |
| Water_Intake (liters) | Daily water intake during workouts |
| Workout_Frequency (days/week) | Number of workout sessions per week |
| Experience_Level | Level of experience, from beginner (1) to expert (3) |
| BMI | Body Mass Index, calculated from height and weight |

```python
#define list of numerical_var
numerical_var = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
# Exercise 3
# Calculate Minimum of numerical variable and explain the result

min_values = df[numerical_var].min()
min_values
```

Min shows the smallest value. Helps spot errors or set a baseline.

Example: Youngest member is 18, lowest calories is 303.

```python
# Exercise 4
# Calculate Maximum of numerical variable and explain the result

max_values = df[numerical_var].max()
max_values
```

Max shows the largest value. Big gap between min and max means wide spread.

Example: Age goes from 18 to 59, Calories from 303 to 1783.

```python
# Exercise 5
# Calculate mean of numerical variable and explain the result


mean_values = df[numerical_var].mean()
mean_values

```

Mean is the average. It can be pulled by extreme values (outliers).

Example: Average calories is ~905, average age is ~39.

```python
# Exercise 6
# Calculate Median of numerical variable and explain the result

median_value = df[numerical_var].median()
median_value



```

Median is the middle value. If mean ≈ median, data is balanced. If not, data is skewed.

Example: Median calories (893) ≈ mean (905), so fairly balanced.

```python
# Exercise 7
# Calculate Mode of numerical variable and explain the result

mode_values = df[numerical_var].mode()
mode_values
```

Mode is the most common value.

Example: Most common experience level is 2, most common age is 43.

```python
# Exercise 8
# Calculate Variance of numerical variable and explain the result

var_values = df[numerical_var].var()
var_values


```

Variance shows how spread out the data is. Higher = more spread. Hard to read because it's squared.

Example: Calories has high variance (74,333), Height has low (0.016).

```python
# Exercise 9
# Calculate Standard Deviation of numerical variable and explain the result


std_values = df[numerical_var].std()
std_values

```

SD (Standard Deviation) is like variance but easier to read. It uses the same units as the data.

Example: Calories SD=273 means most members burn 632-1178 calories (905±273).

```python
# Exercise 10
# Calculate Q1, Q2, Q3 of numerical variable and explain the result

quartiles = df[numerical_var].quantile([0.25, 0.50, 0.75])
quartiles
```

Quartiles split data into 4 parts. Q1-Q3 holds the middle 50%.

Example: Middle 50% of members burn 720-1076 calories.

```python
# Exercise 11
# plot Boxplot and explain result 

# Group variables by similar value ranges
# Becase some variables have very different ranges, we will group them for better visualization
# Else we will not get the insight from all of them 
calories = ['Calories_Burned']
bpm_vars = ['Max_BPM', 'Avg_BPM', 'Resting_BPM']
medium_range = ['Age', 'Weight (kg)', 'Fat_Percentage', 'BMI']
small_range = ['Height (m)', 'Session_Duration (hours)', 'Water_Intake (liters)', 'Workout_Frequency (days/week)', 'Experience_Level']

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Plot 1: Calories Burned (standalone)
sns.boxplot(data=df[calories], ax=axes[0])
axes[0].set_title('Calories Burned')
axes[0].tick_params(axis='x', rotation=45)

# Plot 2: BPM variables (100s-200s)
sns.boxplot(data=df[bpm_vars], ax=axes[1])
axes[1].set_title('Heart Rate Variables (BPM)')
axes[1].tick_params(axis='x', rotation=45)

# Plot 3: Medium range variables (10s-100s)
sns.boxplot(data=df[medium_range], ax=axes[2])
axes[2].set_title('Medium Range (Age, Weight, Fat%, BMI)')
axes[2].tick_params(axis='x', rotation=45)

# Plot 4: Small range variables (0-10)
sns.boxplot(data=df[small_range], ax=axes[3])
axes[3].set_title('Small Range (Height, Duration, etc.)')
axes[3].tick_params(axis='x', rotation=45)

plt.suptitle('Boxplots of Numerical Variables (Grouped by Range)', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

Variables are grouped into 4 plots by similar ranges for better visualization:

- **Plot 1 - Calories Burned:** Standalone because its range is much larger than others. Shows some high outliers.
- **Plot 2 - Heart Rate (BPM):** Groups all heart rate variables together. Max_BPM is highest, Resting_BPM is lowest.
- **Plot 3 - Medium Range:** Age, Weight, Fat%, and BMI have similar scales. Weight and BMI show some outliers.
- **Plot 4 - Small Range:** Variables with values between 0-5. Height is the most consistent (smallest box).

```python
# Exercise 12
# calculate IQR and explain 

Q1 = df[numerical_var].quantile(0.25)
Q3 = df[numerical_var].quantile(0.75)
IQR = Q3 - Q1
print("IQR (Interquartile Range):")
print(IQR)
```

IQR = Q3 - Q1. It shows the spread of the middle 50%. Used to find outliers.

Example: Calories IQR=356 means middle 50% spans 356 calories.

```python
# Exercise 13
# calculate lower_bound of outlier and upper_bound of outlier and explain the result 

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("Lower Bound for Outliers:")
print(lower_bound)
print("\nUpper Bound for Outliers:")
print(upper_bound)
```

Outliers are values too far from the middle (1.5×IQR rule). They need checking.

Example: Calories >1610 are outliers. Max is 1783, so some exist.

```python
# Exercise 14
# calculate Coefficient of Variation of numerical variable and explain the result
# When CV should use?

# CV = (Standard Deviation / Mean) * 100
CV = (df[numerical_var].std() / df[numerical_var].mean()) * 100
print("Coefficient of Variation (%)")
print(CV)
print("\nCV is used to compare variability between datasets with different units or scales.")
```

CV compares spread across different units. Higher CV = more spread.

Example: Height CV is low (~7%), Calories CV is higher (~30%).

```python
# Exercise 15
# Calculate covariance of two numerical variable and explain the result 

# Example: Covariance between Calories_Burned and Session_Duration
covariance = df[['Calories_Burned', 'Session_Duration (hours)']].cov()
print("Covariance Matrix:")
print(covariance)
```

Covariance shows if two things move together (+), opposite (-), or not at all (0).

Example: Longer workouts = more calories burned (positive covariance).

```python
# Exercise 16
# Calculate Correlation of two numerical variable and explain the result 
# What is the difference between correlation and covariance?

correlation = df[numerical_var].corr()

# Heatmap for better visualization
plt.figure(figsize=(12, 10))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

Correlation is like covariance but scaled from -1 to 1. Easier to compare.

Example: 0.8 = strong link, 0.2 = weak link, -0.5 = opposite link.

```python
# Exercise 17
# calculate Skewness of numerical variable and explain the result 

skewness = df[numerical_var].skew()
print("Skewness:")
print(skewness)
```

Skewness shows if data leans left or right. ~0 = balanced, >0 = right tail, <0 = left tail.

Example: Positive skew means most values are low, few are very high.

```python
# Exercise 18
# Plot Distribution of numerical variable and explain the result 

# Example: Distribution of Age
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Distribution of Age')

plt.subplot(1, 2, 2)
sns.histplot(df['Calories_Burned'], kde=True, bins=30, color='orange')
plt.title('Distribution of Calories Burned')
plt.tight_layout()
plt.show()
```

Histogram shows the shape of data:
- **Bell curve (normal)**: Most values in the middle, fewer at edges. Good for statistics.
- **Flat (uniform)**: All values equally common. No typical value.
- **Two peaks (bimodal)**: Two groups in the data. May need to split them.
- **Skewed**: Most values on one side, long tail on the other.

Example: Age looks flat (all ages equally common), Calories looks like a bell curve (most burn around average).

```python
# Exercise 19 
# Calculate Kurtosis  of numerical variable and explain the result 

kurtosis = df[numerical_var].kurtosis()
print("Kurtosis:")
print(kurtosis)
```

Kurtosis shows if data has heavy or light tails:
- **Positive (>0)**: Heavy tails, more outliers than normal
- **Zero (~0)**: Normal distribution, like a bell curve
- **Negative (<0)**: Light tails, fewer outliers than normal

Example: Positive kurtosis means extreme values (very high or very low) appear more often.

```python
# Exercise 20
# calculate Z-score of numerical variable

from scipy import stats

# Calculate Z-scores for all numerical variables
z_scores = df[numerical_var].apply(stats.zscore)
print("Z-scores (first 5 rows):")
print(z_scores.head())
```

Z-score tells how far a value is from average.

Example: Z-score of 2.5 for Calories = burns much more than average.

```python
# Exercise 21
# Identify outlier based on Z-score and print the result 

# Outliers are typically defined as Z-score > 3 or < -3
outliers = (z_scores.abs() > 3).sum()
print("Number of outliers per variable (Z-score > 3):")
print(outliers)
print(f"\nTotal outliers detected: {outliers.sum()}")
```

Z-score outliers are values where |Z| > 3 (very far from average).

Example: 5 outliers in BMI means 5 members have extreme BMI.

```python
# Exercise 22
# Count class of categories vaiable 

categorical_var = df.select_dtypes(include=['str']).columns.tolist()
for col in categorical_var:
    print(f"\n{col}:")
    print(df[col].value_counts())
```

Value counts show how many of each category. Helps check if data is balanced.

Example: Male=500, Female=473 means fairly balanced.

```python
# Exercise 23
# Plot the value counts of each class in the categorical variable

categorical_var = df.select_dtypes(include=['str']).columns.tolist()
fig, axes = plt.subplots(1, len(categorical_var), figsize=(15, 5))

for i, col in enumerate(categorical_var):
    ax = axes[i] if len(categorical_var) > 1 else axes
    df[col].value_counts().plot(kind='bar', ax=ax, color='skyblue')
    ax.set_title(f'{col} Distribution')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Bar plot shows category counts visually. Equal bars = balanced, unequal = imbalanced.

Example: Similar bar heights for workout types = well-balanced data.
